```python
import pandas as pd
import numpy as np
```

```python
cap = pd.read_csv('DC_Capacity.dat', header=1, sep='[')
cap.columns = ['DC', 'cap']
cap['cap'] = cap['cap'].str.rstrip(' ]')
cap['DC'] = cap['DC'].str.rstrip()
cap['DC'] = cap['DC'].str[2:-2]
cap=cap[:-1]
cap['cap'] = cap['cap'].astype(int)
cap
```

```python
fixedCost = pd.read_csv('DC_FixedCost.dat', header=1, sep='[')
fixedCost.columns = ['DC', 'fixedCost']
fixedCost['fixedCost'] = fixedCost['fixedCost'].str.rstrip(' ]')
fixedCost['DC'] = fixedCost['DC'].str.rstrip()
fixedCost['DC'] = fixedCost['DC'].str[2:-2]
fixedCost=fixedCost[:-1]
fixedCost['fixedCost'] = fixedCost['fixedCost'].astype(int)
fixedCost
```

```python
varCost = pd.read_csv('DC_VariableCost.dat', header=1, sep='[')
varCost.columns = ['DC', 'varCost']
varCost['varCost'] = varCost['varCost'].str.rstrip('] ')
varCost['DC'] = varCost['DC'].str.rstrip()
varCost['DC'] = varCost['DC'].str[2:-2]
varCost=varCost[:-1]
varCost['varCost'] = varCost['varCost'].astype(float)
varCost
```

```python
prod = pd.read_csv('Sku_weight_volume.dat', header=1, sep='[')
prod.columns = ['Product', 'Split']
prod = pd.concat([prod[['Product']], prod['Split'].str.split(', ', expand=True)], axis=1)
prod.columns = ['Product', 'Weight', 'Cubic', 'Price']
prod['Price'] = prod['Price'].str.rstrip(']')
prod['Product'] = prod['Product'].str[2:-2]
prod=prod[:-1]
prod['Product'] = prod['Product'].astype(int)
prod['Weight'] = prod['Weight'].astype(float)
prod['Cubic'] = prod['Cubic'].astype(float)
prod['Price'] = prod['Price'].astype(int)
prod
```

```python
custDem = pd.read_csv('Customer_Demand.dat', header=1, sep='[')
custDem.columns = ['Split', 'Quantity']
custDem = pd.concat([custDem['Split'].str.split(' ', expand=True), custDem[['Quantity']]], axis=1)
custDem.columns = ['CID', 'Product', 'Drop','Quantity']
custDem['CID'] = custDem['CID'].str.rstrip()
custDem['Product'] = custDem['Product'].str[1:-2]
custDem['CID'] = custDem['CID'].str[2:-1]
custDem['Quantity'] = custDem['Quantity'].str[:-1]
custDem = custDem[:-1]
custDem = custDem.drop(['Drop'], axis=1)
custDem['Product'] = custDem['Product'].astype(int)
custDem['Quantity'] = custDem['Quantity'].astype(int)
custDem
```

```python
transpo = pd.read_csv('Site_lat_long.dat', header=0, sep='\t')
transpo =transpo.reset_index()
transpo.columns = ['CID', 'Latitude', 'Longitude']
transpo['CID'] = transpo['CID'].str.rstrip()
transpo['Latitude'] = transpo['Latitude'].astype(float)
transpo['Longitude'] = transpo['Longitude'].astype(float)
transpo1 = transpo[7:-1]
transpo1 = transpo1.reset_index(drop=True)
transpo1
```

```python
hi=transpo2.set_index(['DC']).unstack()
transpo2
```

```python
mo['lat1']=hi.iloc[0]
mo['long1']=hi.iloc[-7]
mo['lat2']=hi.iloc[1]
mo['long2']=hi.iloc[-6]
mo['lat3']=hi.iloc[2]
mo['long3']=hi.iloc[-5]
mo['lat4']=hi.iloc[3]
mo['long4']=hi.iloc[-4]
mo['lat5']=hi.iloc[4]
mo['long5']=hi.iloc[-3]
mo['lat6']=hi.iloc[5]
mo['long6']=hi.iloc[-2]
mo['lat7']=hi.iloc[6]
mo['long7']=hi.iloc[-1]
```

```python
mo
```

```python
mo['1']=((mo['Longitude']-mo['long1'])**2 + (mo['Latitude']-mo['lat1'])**2)**(1/2)
mo['2']=((mo['Longitude']-mo['long2'])**2 + (mo['Latitude']-mo['lat2'])**2)**(1/2)
mo['3']=((mo['Longitude']-mo['long3'])**2 + (mo['Latitude']-mo['lat3'])**2)**(1/2)
mo['4']=((mo['Longitude']-mo['long4'])**2 + (mo['Latitude']-mo['lat4'])**2)**(1/2)
mo['5']=((mo['Longitude']-mo['long5'])**2 + (mo['Latitude']-mo['lat5'])**2)**(1/2)
mo['6']=((mo['Longitude']-mo['long6'])**2 + (mo['Latitude']-mo['lat6'])**2)**(1/2)
mo['7']=((mo['Longitude']-mo['long7'])**2 + (mo['Latitude']-mo['lat7'])**2)**(1/2)
mo
```

```python
mo['min']=pd.concat([mo['1'], mo['2'], mo['3'], mo['4'], mo['5'], mo['6'], mo['7']], axis=1).min(axis=1)
```

```python
mo
```

```python
#CHAPTER 8
mo['k'] = 0
mo['k'][mo['Weight']*8 < 12] = 32
mo['k'][mo['Weight']*8 == 12] = 28
mo['k'][(mo['Weight']*8 > 12) & (mo['Weight']*8 < 60)] = 28
mo['k'][mo['Weight']*8 == 60] = 23
mo['k'][(mo['Weight']*8 > 60) & (mo['Weight']*8 < 300)] = 23
mo['k'][mo['Weight']*8 == 300] = 16
mo['k'][(mo['Weight']*8 > 300) & (mo['Weight']*8 < 900)] = 16
mo['k'][mo['Weight']*8 == 900] = 8
mo['k'][mo['Weight']*8 > 900] = 8
```

```python
mo.sort_values('k')
```

```python
mo['tCost']=mo['min']/mo['k']*mo['Quantity']
mo.sort_values('tCost')
```

```python
ih=varCost.set_index(['DC']).unstack()
ih
```

```python
mo['vCost'] = 0
mo['vCost'][mo['1'] == mo['min']] = mo['Quantity']*ih.iloc[0]
mo['vCost'][mo['2'] == mo['min']] = mo['Quantity']*ih.iloc[1]
mo['vCost'][mo['3'] == mo['min']] = mo['Quantity']*ih.iloc[2]
mo['vCost'][mo['4'] == mo['min']] = mo['Quantity']*ih.iloc[3]
mo['vCost'][mo['5'] == mo['min']] = mo['Quantity']*ih.iloc[4]
mo['vCost'][mo['6'] == mo['min']] = mo['Quantity']*ih.iloc[5]
mo['vCost'][mo['7'] == mo['min']] = mo['Quantity']*ih.iloc[6]
```

```python
po['tCost1']=po['1']/po['k']*po['Quantity']
po['tCost2']=po['2']/po['k']*po['Quantity']
po['tCost3']=po['3']/po['k']*po['Quantity']
po['tCost4']=po['4']/po['k']*po['Quantity']
po['tCost5']=po['5']/po['k']*po['Quantity']
po['tCost6']=po['6']/po['k']*po['Quantity']
po['tCost7']=po['7']/po['k']*po['Quantity']
po['vCost1'] = po['Quantity']*ih.iloc[0]
po['vCost2'] = po['Quantity']*ih.iloc[1]
po['vCost3'] = po['Quantity']*ih.iloc[2]
po['vCost4'] = po['Quantity']*ih.iloc[3]
po['vCost5'] = po['Quantity']*ih.iloc[4]
po['vCost6'] = po['Quantity']*ih.iloc[5]
po['vCost7'] = po['Quantity']*ih.iloc[6]
po['tv1'] = po['tCost1'] + po['vCost1']
po['tv2'] = po['tCost2'] + po['vCost2']
po['tv3'] = po['tCost3'] + po['vCost3']
po['tv4'] = po['tCost4'] + po['vCost4']
po['tv5'] = po['tCost5'] + po['vCost5']
po['tv6'] = po['tCost6'] + po['vCost6']
po['tv7'] = po['tCost7'] + po['vCost7']
po['min'] = pd.concat([po['tv1'], po['tv2'], po['tv3'], po['tv4'], po['tv5'], po['tv6'], po['tv7']], axis=1).min(axis=1)
po
```
